```python
import pandas as pd

# Load the file
data = pd.read_csv('AAPL.csv')

# Create a new column for the price range
# data['Interval_Range'] = "[" + data['Low'].astype(str) + "-" + data['High'].astype(str) + "]"
# df = data.drop(columns=['Open','High','Low','Close','Adj Close','Volume'])

data['Interval_Price'] = data.apply(lambda row: pd.Interval(row['Low'], row['High'], closed='both'), axis=1)
data['midpoint'] = (data['Low'] + data['High']) / 2
df = data.drop(columns=['Open','High','Low','Close','Adj Close','Volume'])
# Save the data to a new CSV file
df.to_csv('Interval_Data.csv', index=False)
```

```python
import pandas as pd
from statsmodels.tsa.vector_ar.var_model import VAR
from sklearn.model_selection import train_test_split

# Load the data
df = pd.read_csv('Interval_Data.csv', index_col='Date', parse_dates=True)

# Check the data
df.head()
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tools.eval_measures import rmse

# Extract the `High` and `Low` series
df = df[['High', 'Low']]

# Check for stationarity
def adf_test(series, signif=0.05):
    """Perform ADFuller to test for Stationarity of given series and print report"""
    r = adfuller(series, autolag='AIC')
    output = {'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
    p_value = output['pvalue']
    return p_value

# Apply ADF test to each column
for name, column in df.iteritems():
    adf_test(column)

# Split the data
nobs = 12
df_train, df_test = df[0:-nobs], df[-nobs:]

# Fit the model
model = VAR(df_train)
model_fitted = model.fit()

# Make predictions
df_forecast = model_fitted.forecast(y=df_train.values[-model_fitted.k_ar:], steps=nobs)

# Create a DataFrame for the forecast values
df_forecast = pd.DataFrame(df_forecast, index=df.index[-nobs:], columns=df.columns+'_forecast')

# Calculate RMSE
for col in df.columns:
    print('RMSE for {}: {}'.format(col, rmse(df_test[col], df_forecast[col+'_forecast'])))

# Print the forecast
df_forecast

```

```python
import matplotlib.pyplot as plt

# Create a combined DataFrame of actual and forecast values
df_combined = pd.concat([df_test, df_forecast], axis=1)

# Plot actual vs forecast values
for col in df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(df_combined[col], label='Actual')
    plt.plot(df_combined[col+'_forecast'], label='Forecast')
    plt.title(col + ': Forecast vs Actuals')
    plt.legend()
    plt.show()
```
